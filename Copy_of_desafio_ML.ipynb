# Variáveis 

Dependente: depende de outras variáveis para existir;

Independente: não depende de outras variáveis para existir;


Contínua: são variáveis numéricas com um número infinito de valores;

Categórica: são variáveis que contem um número finito de categorias; 

# Definição do tipo de problema

Regressão - quando alguma feature é dependente e a variável for contínua;

Classificação - quando alguma feature é dependente e a variável for categórica;

Clustering - quando nenhuma feature que será utilizada é dependente;

```
def list_to_str(x):  
    strg = " "      
    return (strg.join(x))              
```

```
df['colors'] = df['colors'].map(list_to_str)  
df['categories'] = df['categories'].map(list_to_str)  

df
```

```
ProfileReport(df)
```

```
df.drop(['brand', 'isSale'], axis=1, inplace=True)
```

```
df
```

```
#encoder utilizando a função onehot
encoder = OneHotEncoder(cols=['categories', 'colors', 'condition'], use_cat_names=True)

encoder.fit(df)
```

```
df = encoder.transform(df)
```

```
df
```

```
x = df.iloc[:, 1:404]
x
```

```
#Utilizando o algoritmo com KMeans, para 4 clusters e random_state 1
kmeans = KMeans(n_clusters=4, random_state =0)
kmeans.fit(x)
```

```
#Descobrindo os valores dos centroides
kmeans.cluster_centers_
```

```
#Distância por linha em relação aos 4 clusters
distance = kmeans.fit_transform(x)
distance
```

```
#
labels = kmeans.labels_
labels
```

```
resultado = pd.Series(labels, index=df['id'], name='cluster_id')
resultado
```
